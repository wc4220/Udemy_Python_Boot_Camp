```python
# Method 2 - More robust method, utilising the loop approach

import string
import random

letters = list(string.ascii_letters)     # lowercase + uppercase combined
numbers = list(string.digits)            # ['0', '1', ... '9']
symbols = list(string.punctuation)       # all common symbols


print("Welcome to the PyPassword Generator!")
nr_letters = int(input(f"How many letters would you like in your password? \n"))
nr_symbols = int(input(f"How many symbols would you like? \n"))
nr_numbers = int(input(f"How many numbers would you like? \n"))

# Create an empty list to store choices

let_list = []
sym_list = []
num_list = []

# Repeat the for loop number of types specified by the user

for l in range(nr_letters):                  # Specify the number of times the loop needs to be repeated.
    let_list.append(random.choice(letters))  # Append the chosen random variable into the empty list.

for s in range(nr_symbols):
    sym_list.append(random.choice(symbols))

for n in range(nr_numbers):
    num_list.append(random.choice(numbers))

# Combine the selected list into one
combined_list = let_list + sym_list + num_list

# Shuffle the list
random.shuffle(combined_list)

# Takes a list of strings and combines them into one single string
print(f"Your password is: {"".join(combined_list)} ")
```
