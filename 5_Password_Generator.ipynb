```python
# First Exercise - Finding the maximum

# Option 1 - Using a loop to find the maximum score

student_scores = [88, 92, 79, 93, 85]
highest_score = student_scores[0]  # Initialize with the first score

for score in student_scores:
    if score > highest_score:
        highest_score = score

print(highest_score) # This will need to be outside of the loop to avoid printing multiple times

### Option 2 - Using the built-in max() function
max_score = max(student_scores)
print(f"The highest score in the class is: {max_score}")
```

```python
# Second Exercise - Solving the Gauss Challenge

sum_of_numbers = 0 # Initialise (This should be outside of the loop)

for number in range(1, 101):
    sum_of_numbers = number + sum_of_numbers

print(sum_of_numbers)
```

```python
# Third Exercise - FizzBuzz

for number in range (1,101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 5 == 0:
        print("Buzz")
    elif number % 3 == 0:
        print("Fizz")
    else:
        print(number)
```
